```python

class Screen:   
    
    def __init__(self):
        
        self.width = 0 #1000
        self.height =0 #1000
        self.x = 0   #500
        self.y = 0   #550

        
        self.napis = '' #'Gra -> Warcaby'
        
    
    #miejsce pojawienia sie okna z grą
    def umiesc_okno(self, x, y):
    
        self.x = x
        self.y = y
        os.environ['SDL_VIDEO_WINDOW_POS'] = "%d,%d" % (self.x,self.y)

    #rozmiar i podpis okna
    def nadaj_rozmiar(self, width, height):
        
        self.width = width
        self.height = height
        
        screen = pygame.display.set_mode(self.width, self.height)
        
    #nazwa okienka
    def nadaj_nazwe(self, napis):
    
        self.napis = napis
        pygame.display.set_caption(napis)
        
        

        #definicja czcionki i rozmiar
        #smallfont = pygame.font.SysFont('Arial',80) 
        #text1 = smallfont.render('B' , True , color_white) 
        #text2 = smallfont.render('C' , True , color_white) 


        
        
window = Screen()
window.umiesc_okno(500, 550)
```
